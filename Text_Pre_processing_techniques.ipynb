```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Download datasets
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')

# Input paragraphs
text = """
Reading books is a wonderful habit! Many people enjoy reading stories, novels, and articles. However, some readers find long books difficult; they prefer short stories or magazines. Good books always teach valuable lessons and improve our knowledge.

Healthy habits are important for a good life. Doctors often say: eat healthy food, exercise regularly, and sleep properly. People who follow healthy routines usually feel energetic, active, and happy!
"""

print("Original Text:\n", text)

# 1. Remove punctuation
text_no_punct = text.translate(str.maketrans('', '', string.punctuation))

print("\nText after removing punctuation:\n", text_no_punct)

# 2. Tokenization
tokens = word_tokenize(text_no_punct.lower())

# 3. Stopword removal
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in tokens if word not in stop_words]

print("\nAfter Stopword Removal:\n", filtered_words)

# 4. Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]

print("\nAfter Stemming:\n", stemmed_words)

# 5. Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

print("\nAfter Lemmatization:\n", lemmatized_words)
```
