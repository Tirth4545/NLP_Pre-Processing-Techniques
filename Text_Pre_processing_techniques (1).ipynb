```python

import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize, sent_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')

# Input paragraphs
text = """
Reading books is a wonderful habit! Many people enjoy reading stories, novels, and articles. However, some readers find long books difficult; they prefer short stories or magazines. Good books always teach valuable lessons and improve our knowledge.

Healthy habits are important for a good life. Doctors often say: eat healthy food, exercise regularly, and sleep properly. People who follow healthy routines usually feel energetic, active, and happy!
"""

print("Original Text:\n", text)

# 1. Remove punctuation
text_no_punct = text.translate(str.maketrans('', '', string.punctuation))
print("\nText after removing punctuation:\n", text_no_punct)

# 2. Tokenization
tokens = word_tokenize(text_no_punct.lower())

# 3. Stopword removal
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in tokens if word not in stop_words]
print("\nAfter Stopword Removal:\n", filtered_words)

# 4. Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]
print("\nAfter Stemming:\n", stemmed_words)

# 5. Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print("\nAfter Lemmatization:\n", lemmatized_words)


# ================================
# ASSIGNMENT 2 – FEATURE ENGINEERING
# ================================

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Convert text into list format
documents = [text]

# 1. Bag of Words
bow = CountVectorizer()
bow_matrix = bow.fit_transform(documents)

print("\nBoW Feature Names:\n", bow.get_feature_names_out())
print("\nBoW Matrix:\n", bow_matrix.toarray())

# 2. TF-IDF
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(documents)

print("\nTF-IDF Feature Names:\n", tfidf.get_feature_names_out())
print("\nTF-IDF Matrix:\n", tfidf_matrix.toarray())


# ASSIGNMENT 2 – WORD2VEC

# Install gensim (run once if needed)
!pip install gensim

from gensim.models import Word2Vec

# New text (200+ words)
text_w2v = """
Student life is one of the most important phases of a person’s life. During this time, students learn new skills, gain knowledge, and build their future. School and college provide opportunities for education as well as personal growth. Students attend classes, complete assignments, and prepare for exams to achieve good results.

Apart from studies, student life also includes extracurricular activities such as sports, music, and cultural events. These activities help students develop confidence, teamwork, and leadership skills. Friends play a very important role in student life, as they support each other and share experiences together.

Time management is an essential skill for students. Balancing studies, hobbies, and social life can be challenging, but it teaches responsibility and discipline. Teachers guide students in their academic journey and motivate them to achieve their goals.

In today’s world, technology has become a major part of student life. Students use the internet, online learning platforms, and digital tools to enhance their knowledge. E-learning and virtual classrooms have made education more accessible and flexible.

Overall, student life is a time of learning, growth, and self-discovery. The habits and values developed during this period help shape a successful and meaningful future.
"""

# Sentence tokenization
sentences = sent_tokenize(text_w2v)

# Word tokenization
data = [word_tokenize(sentence.lower()) for sentence in sentences]

print("\nSample Tokenized Sentences:\n", data[:2])

# Train Word2Vec model
model = Word2Vec(
    sentences=data,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4
)

# Vocabulary
print("\nVocabulary:\n", list(model.wv.index_to_key))

# Similarity checks
print("\nSimilarity between 'technology' and 'learning':", model.wv.similarity('technology', 'learning'))
print("\nSimilarity between 'Student' and 'Education':", model.wv.similarity('student', 'education'))
print("\nSimilarity between 'life' and 'Learning':", model.wv.similarity('life', 'learning'))
print("\nSimilarity between 'Future' and 'Student':", model.wv.similarity('future', 'student'))
print("\nSimilarity between 'Teachers' and 'Technology':", model.wv.similarity('teachers', 'technology'))
print("\nSimilarity between 'Teachers' and 'Student':", model.wv.similarity('teachers', 'student'))

# Most similar words
print("\nWords similar to 'technology':")
print(model.wv.most_similar('technology'))


# ================================
# CONCLUSION
# ================================

print("\nConclusion:")
print("1. Bag of Words converts text into frequency-based vectors.")
print("2. TF-IDF highlights important words.")
print("3. Word2Vec captures semantic relationships between words.")
```
